```python
import pandas as pd
import numpy as np
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('oligo_table_label_all_with_refalt.csv' )
print(df.shape)
df.head()
```

Filtering of df by set names

```python
np.unique(df['set_name'])
```

```python
RM_var = df.loc[df['set_name'] == 'BY_RM_Variants']
Random_DNA_guide = df.loc[df['set_name']== 'Random_DNA_guide']
inviable_gene_KO =  df.loc[df['set_name']== 'inviable_gene_KO']
inviable_gene_synonymous = df.loc[df['set_name']== 'inviable_gene_synonymous']
```

```python
RM_var_dropNans_DF = RM_var.dropna(subset=('ind1_coor_effect_direction','ind1_coor_pass_fdr','ind2_coor_effect_direction',
                                  'ind2_coor_pass_fdr'), how= 'all')
RM_var_dropNans_DF.head()
```

# V1 of df preprocessing

DF : 
- remove all rols with nans
- T & T  = 1 
- T & F or F & T = 0


```python
#getting index of T/F or F/T in order to make label = 0 
TF_RM_var = (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == False ) &  (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == True) | (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == False ) & (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == False)|  (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == True ) & (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == False)
#print('head' , RM_var_dropNans_DF[TF_RM_var].head())
RM_var_dropNans_DF_TF = RM_var_dropNans_DF[TF_RM_var]
RM_var_dropNans_DF_TF['Label'] = 0 
#TF_RM_var['Label'] = 0
RM_var_dropNans_DF_TF.shape
```

```python
#now getting index for TT in order to make label = 1 
RM_TT = (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == True) &  (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == True)
RM_var_dropNans_DF_TT = RM_var_dropNans_DF[RM_TT]
RM_var_dropNans_DF_TT['Label'] = 1
RM_var_dropNans_DF_TT.head()
```

```python
RM_var_dropNans_DF_TT.shape
```

```python
#concate both df of labels 
RM_var_DF_Complete = pd.concat((RM_var_dropNans_DF_TT, RM_var_dropNans_DF_TF))
RM_var_DF_Complete.head()
```

```python
labels = np.array(RM_var_DF_Complete['Label'])
#np.save('RM_var_DF.npy', labels)
#pd.DataFrame.to_csv('RM_var_DF_Complete.csv', RM_var_DF_Complete)
```

# V2 of df preprocessing

- both true & -1 = 1 
- rest (F/T or T/F or F/F) = 0 


still starting with dropping all NaNs

```python
condition_index = (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == True) &  (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == True)  & (RM_var_dropNans_DF['ind1_coor_effect_direction'] == -1) & (RM_var_dropNans_DF['ind2_coor_effect_direction'] == -1)
```

```python
new_df = RM_var_dropNans_DF[condition_index]
print(new_df.shape)
#new_df.head()
new_df['Label'] = 1
new_df.head()
```

```python
sum(new_df['Label']==1)
```

```python
other_conds_index = (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == False ) &  (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == False)| (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == False ) &  (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == True)| (RM_var_dropNans_DF['ind1_coor_pass_fdr'] == True ) & (RM_var_dropNans_DF['ind2_coor_pass_fdr'] == False)
other_df = RM_var_dropNans_DF[other_conds_index]
print(other_df.shape)
other_df['Label'] = 0
other_df.head()
```

```python
#concate both df of labels 

RM_var_DF_Complete_2 = pd.concat((new_df, other_df))
print(RM_var_DF_Complete_2.shape)
RM_var_DF_Complete_2.head()
```

```python
labels = np.array(RM_var_DF_Complete_2['Label'])
np.save('RM_var_DF_V3_518.npy', labels)
labels.shape
```

# extracting all features from new DF -- V2

```python
#donor seq & ogoligo seq
RM_var_noPam = np.array( RM_var_DF_Complete_2['guide_noPAM'])
RM_var_donor_seq = np.array( RM_var_DF_Complete_2['donor_seq'])
RM_var_edit_guide = np.array(RM_var_DF_Complete_2['edit_pos_in_guide'])

print('no pam', RM_var_noPam[0])
print('donor', RM_var_donor_seq[0])
print('guide', RM_var_edit_guide[0])
```

#  DF v3

- ref 
- alt 
- pos 

```python
RM_ref_seq = np.array( RM_var_DF_Complete_2['ref'])
RM_alt_seq = np.array( RM_var_DF_Complete_2['alt'])
RM_pos_guide = np.array( RM_var_DF_Complete_2['edit_pos_in_guide'])
print('RM_ref_seq ', RM_ref_seq[0])
print('RM_alt_seq ', RM_alt_seq[0])
print('RM_pos_guide ', RM_pos_guide[0])
```

# One Hot Encoding

```python
# integer encode
noPam_label_encoder = LabelEncoder()
noPam_integer_encoded = noPam_label_encoder.fit_transform(RM_var_noPam)
print(noPam_integer_encoded)

donorSeq_label_encoder = LabelEncoder()
donorSeq_integer_encoded = donorSeq_label_encoder.fit_transform(RM_var_donor_seq)
print(donorSeq_integer_encoded)

guide_label_encoder = LabelEncoder()
guide_integer_encoded = guide_label_encoder.fit_transform(RM_var_edit_guide)
print(guide_integer_encoded)

################ V3 

RM_ref_seq_encoder = LabelEncoder()
RM_ref_seq_integer_encoded = RM_ref_seq_encoder.fit_transform(RM_ref_seq)
print(RM_ref_seq_integer_encoded)

RM_pos_guide_encoder = LabelEncoder()
RM_pos_guide_integer_encoded = RM_pos_guide_encoder.fit_transform(RM_pos_guide)
print(RM_pos_guide_integer_encoded)

RM_alt_seq_encoder = LabelEncoder()
RM_alt_seq_integer_encoded = RM_alt_seq_encoder.fit_transform(RM_alt_seq)
print(RM_alt_seq_integer_encoded)
```

```python
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)

noPam_integer_encoded = noPam_integer_encoded.reshape(len(noPam_integer_encoded), 1)
noPam_onehot_encoded = onehot_encoder.fit_transform(noPam_integer_encoded)
print(noPam_onehot_encoded[0:1])

donorSeq_integer_encoded = donorSeq_integer_encoded.reshape(len(noPam_integer_encoded), 1)
donorSeq_onehot_encoded = onehot_encoder.fit_transform(donorSeq_integer_encoded)
print(noPam_onehot_encoded[0:1])

guide_integer_encoded = guide_integer_encoded.reshape(len(guide_integer_encoded), 1)
guide_onehot_encoded = onehot_encoder.fit_transform(guide_integer_encoded)
print(guide_onehot_encoded[0:1])

############# V3 
RM_alt_seq_integer_encoded = RM_alt_seq_integer_encoded.reshape(len(RM_alt_seq_integer_encoded), 1)
RM_alt_seq_onehot_encoded = onehot_encoder.fit_transform(RM_alt_seq_integer_encoded)
print(RM_alt_seq_onehot_encoded[0:1])

RM_pos_guide_integer_encoded = RM_pos_guide_integer_encoded.reshape(len(RM_pos_guide_integer_encoded), 1)
RM_pos_guide_onehot_encoded = onehot_encoder.fit_transform(RM_pos_guide_integer_encoded)
print(RM_pos_guide_onehot_encoded[0:1])

RM_ref_seq_integer_encoded = RM_ref_seq_integer_encoded.reshape(len(RM_ref_seq_integer_encoded), 1)
RM_ref_seq_integer_onehot_encoded = onehot_encoder.fit_transform(RM_ref_seq_integer_encoded)
print(RM_ref_seq_integer_onehot_encoded[0:1])
```

```python
# invert first example
noPam_inverted = noPam_label_encoder.inverse_transform([argmax(noPam_onehot_encoded[0, :])])
print(noPam_inverted)

donorSeq_inverted = donorSeq_label_encoder.inverse_transform([argmax(donorSeq_onehot_encoded[0, :])])
print(donorSeq_inverted)

guide_inverted = guide_label_encoder.inverse_transform([argmax(guide_onehot_encoded[0, :])])
print(guide_inverted)

##### V2 
RM_ref_seq_integer_inverted = RM_ref_seq_encoder.inverse_transform([argmax(RM_ref_seq_integer_onehot_encoded[0, :])])
print(RM_ref_seq_integer_inverted)

RM_pos_guide_inverted = RM_pos_guide_encoder.inverse_transform([argmax(RM_pos_guide_onehot_encoded[0, :])])
print(RM_pos_guide_inverted)

RM_alt_seq_inverted = RM_alt_seq_encoder.inverse_transform([argmax(RM_ref_seq_integer_encoded[0, :])])
print(RM_alt_seq_inverted)
```

```python
np.savez('RM_alt_seq_Complete_V3.npz',  RM_alt_seq, RM_alt_seq_integer_encoded, RM_ref_seq_integer_onehot_encoded)
np.savez('RM_pos_guide_Complete_V3.npz',  RM_pos_guide, RM_pos_guide_integer_encoded, RM_pos_guide_onehot_encoded)
np.savez('RM_ref_seq_Complete_V3.npz',  RM_ref_seq, RM_ref_seq_integer_encoded, RM_ref_seq_integer_onehot_encoded)
```
