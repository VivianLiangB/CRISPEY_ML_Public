```python
'''
Load arrays for preping DF: 

For structure of npz:
arr one is DNA seq, arr two is label encoding, arr three is one hot 

'''
RM_alt_seq = np.load('RM_alt_seq_Complete_V3.npz')
RM_pos_guide = np.load('RM_pos_guide_Complete_V3.npz')
RM_ref_seq = np.load('RM_ref_seq_Complete_V3.npz')

RM_alt_seq_onehot =   RM_alt_seq['arr_2']
RM_pos_guide_onehot = RM_pos_guide['arr_2']
RM_ref_seq_onehot = RM_ref_seq['arr_2']


print('shapes for alt seq, post guide,  ref:',
      RM_alt_seq_onehot.shape, RM_pos_guide_onehot.shape, RM_ref_seq_onehot.shape)

#stack arrays via cols for one hot 
RM_stacks = np.column_stack((RM_alt_seq_onehot, RM_pos_guide_onehot,
                             RM_ref_seq_onehot))
Y_labels= np.load('RM_var_DF_V3.npy')
print('RM stack shape', RM_stacks.shape)
print('y label shape', Y_labels.shape)
```

```python
# Stacking cols via label encoder instead of one hot  -- just pref for it to be easier 
# Extract the values for features and create a list called featuresAll
RM_alt_seq_cat  = RM_alt_seq['arr_1']
RM_pos_guide_cat  = RM_pos_guide['arr_1']
RM_ref_seq_cat  = RM_ref_seq['arr_1']
RM_stack_feats = np.column_stack((RM_alt_seq_cat, RM_pos_guide_cat,
                             RM_ref_seq_cat))
print(RM_stack_feats)
print(RM_stack_feats.shape)
```

```python
#Finding the relationship between ref, alt, guide 
featuresAll = []
feat_alt = []
feat_guide = []

for feature in RM_stack_feats:
    feat_alt.append(RM_stack_feats[:,0]) #alt
    #feat_guide.append(RM_stack_feats[:,1]) # guide 
    featuresAll.append(RM_stack_feats[:,2]) #ref
```

```python
groups = ('Ref','Guide','Alt')
colors = ('blue', 'green','red')
data = ((featuresAll[:7799], feat_alt[:7799]), 
        (featuresAll[7799:15598], feat_alt[7799:15598]), 
        (featuresAll[15598:23397], feat_alt[15598:23397]))

for item, color, group in zip(data,colors,groups): 
   # item = (featuresAll[:7799], feat_alt[:7799]), (featuresAll[7799:15598], feat_alt[7799:15598]), (featuresAll[15598:23397], feat_alt[15598:23397])
    x, y  = item
    plt.scatter(x, y,color=color,alpha=1)
    plt.title(' DF Scatter Plot relationship between ref and alt ')
plt.xlabel('ref')
plt.ylabel('alt')
plt.show()
```
